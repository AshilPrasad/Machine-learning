```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[250,89.5,48.5]]))
```

```python
print("slope is",list(zip(x,model.coef_)))
```

```python
print("intercept is",model.intercept_)
```

```python
df1=pd.DataFrame({"Actual value":y_test,"predicted value":y_pred,"error":y_test-y_pred})
df1
```

```python
#performence evaluation

#mean absoluate error
from sklearn.metrics import mean_absolute_error
print("mean absoluate error is",mean_absolute_error(y_test,y_pred))
```

```python
#mean absolute percentage error
from sklearn.metrics import mean_absolute_percentage_error
print("mean absolute error percenatge is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
# mean squard error
from sklearn.metrics import mean_squared_error
print("mean squard error is",mean_squared_error(y_test,y_pred))
```

```python
#root mean squard error
import numpy as np
rmse=mean_squared_error(y_test,y_pred)
sqd=np.sqrt(rmse)
print("root mean squard error is",sqd)
```

```python
#r2 score  /coefficient of determination
from sklearn.metrics import r2_score
print("r2 score is",r2_score(y_test,y_pred))
```
