```python
x1=df.iloc[:,:-1]
x1
```

```python
y=df.iloc[:,-1]
y
```

```python
#using onehotencoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown="ignore"),["State"]),remainder="passthrough")
x=col_trans.fit_transform(x1)
x
```

```python
#spliting training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({"Actual value":y_test,"predicted value":y_pred,"error":y_test-y_pred})
df1
```

```python
print("slope is",list(zip(x1,model.coef_)))
```

```python
print("intercept is",model.intercept_)
```

```python
#performence evaluation

#mean absoluate error
from sklearn.metrics import mean_absolute_error
print("mean absolute error is",mean_absolute_error(y_test,y_pred))
```

```python
#mean absolute percentage error
from sklearn.metrics import mean_absolute_percentage_error
print("mean absolute percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error
print("mean squared error is",mean_squared_error(y_test,y_pred))
```

```python
#root mean squared error
sqd=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(sqd)
print("root mean squraed error is",rmse)
```

```python
#r2 score or coefficient of determination
from sklearn.metrics import r2_score
print("r2 score is",r2_score(y_test,y_pred))
```
