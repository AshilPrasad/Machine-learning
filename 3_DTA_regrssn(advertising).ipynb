```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
x_test
```

```python
#model creation
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(model,feature_names=['TV','Radio','Newspaper'],class_names=['Sales'],filled=True)
```

```python
#performence evaluation
from sklearn.metrics import mean_absolute_error
print('mean_absolute_error is',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print(' mean_absolute_percentage_error is', mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print('mean_squared_error is',mean_squared_error(y_test,y_pred))
```

```python
#root mean squared error
msr=mean_squared_error(y_test,y_pred)
rmsr=np.sqrt(msr)
print(rmsr)
```

```python
from sklearn.metrics import r2_score
print('r2 score is',r2_score(y_test,y_pred))
```
