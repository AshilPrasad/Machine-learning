```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["movie"]=le.fit_transform(df["movie"])
df["year"]=le.fit_transform(df["year"])
df["certificate"]=le.fit_transform(df["certificate"])
df["genre.1"]=le.fit_transform(df["genre.1"])
df["genre.2"]=le.fit_transform(df["genre.2"])
df["genre.3"]=le.fit_transform(df["genre.3"])
df["metascore"]=le.fit_transform(df["metascore"])
df["time_minute"]=le.fit_transform(df["time_minute"])
df["gross_earning"]=le.fit_transform(df["gross_earning"])
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
x_test
```

```python
#model creation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
KNN_model=KNeighborsClassifier()
NBA_model=MultinomialNB()
SVMA_model=SVC()
lis=[KNN_model,NBA_model,SVMA_model]



```

```python
for i in lis:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(y_pred)
  print("accuracy score:",accuracy_score(y_test,y_pred))
  print("classification_report:",classification_report(y_test,y_pred))
print()
```
