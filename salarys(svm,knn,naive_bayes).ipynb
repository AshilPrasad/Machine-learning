```python
df1=df.drop(columns=['employment_type','salary','salary_currency','remote_ratio'],axis=1)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['experience_level']=le.fit_transform(df1['experience_level'])
df1['job_title']=le.fit_transform(df1['job_title'])
df1['employee_residence']=le.fit_transform(df1['employee_residence'])
df1['company_location']=le.fit_transform(df1['company_location'])
df1['company_size']=le.fit_transform(df1['company_location'])
df1
```

```python
x=df1.drop(['salary_in_usd'],axis=1).values
x
```

```python
y=df1['salary_in_usd'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
x_test
```

```python
#model creation using knn
from sklearn.neighbors import  KNeighborsClassifier
model1=KNeighborsClassifier(n_neighbors=5)
model1.fit(x_train,y_train)
y_pred1=model1.predict(x_test)
y_pred1
```

```python
#performence evalution using knn
```

```python
#model creation using naive bayes algorithm
from sklearn.naive_bayes import GaussianNB
model2=GaussianNB()
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)
y_pred2
```

```python
#model creation using svm
from sklearn.svm import SVC
model3=SVC()
model3.fit(x_train,y_train)
y_pred3=model3.predict(x_test)
y_pred3
```
